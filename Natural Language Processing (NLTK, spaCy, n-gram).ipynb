**NLTK vs spaCy:**
1. Stemming vs Lemmatization:
    - NLTK PorterStemmer cuts words aggressively: animation -> anim
    - WordNet Lemmatizer improves but requires POS to be fully accurate. sat -> sat
    - spaCy produces more natural lemmas: sat -> sit
 2. POS Tagging:
    - NLTK pipeline here does NOT include POS tagging by default.
    - spaCy provides built-in POS tags (NOUN, VERB, ADJ, etc.)
 3. Stopword Handling:
    - Both NLTK and spaCy remove stopwords effectively. spaCy integrates this as token attribute (token.is_stop).
 4. Pipeline Simplicity:
    - NLTK requires multiple steps manually.
    - spaCy provides everything in a single pipeline (tokenization,
      POS tagging, lemmatization).

  **Conclusion**: spaCy is more powerful and context-aware. NLTK gives more control.

```python
#PART2

import numpy as np
import string
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer, WordNetLemmatizer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

# Ensure resources
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

# Initialize tools
stop_words = set(stopwords.words('english'))
punct = set(string.punctuation)
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

print("Libraries loaded.")
```

```python
def preprocess_text(text, method='lemma'):
    """
    Preprocess text:
    lowercase -> tokenize -> remove stopwords/punct -> stem/lemma
    """
    # Step 1: Lowercase
    text = text.lower()

    # Step 2: Tokenize
    tokens = word_tokenize(text)

    # Step 3: Remove stopwords/punctuation + keep alphabetic tokens
    tokens = [
        t for t in tokens
        if t not in stop_words and t not in punct and t.isalpha()
    ]

    # Step 4: Stem or Lemmatize
    if method == 'stem':
        tokens = [stemmer.stem(t) for t in tokens]
    elif method == 'lemma':
        tokens = [lemmatizer.lemmatize(t) for t in tokens]
    else:
        raise ValueError("method must be 'stem' or 'lemma'")

    return ' '.join(tokens)
```

```python
df['processed_stemmed'] = df['text'].apply(
    lambda x: preprocess_text(x, method='stem')
)

df['processed_lemmatized'] = df['text'].apply(
    lambda x: preprocess_text(x, method='lemma')
)

print("Preprocessing done!\n")

# Preview results
print("Original:  ", df['text'].iloc[1][:100])
print("Stemmed:   ", df['processed_stemmed'].iloc[1][:100])
print("Lemmatized:", df['processed_lemmatized'].iloc[1][:100])
```

```python
# Create BoW representation
count_vectorizer = CountVectorizer(max_features=5000)

X_bow = count_vectorizer.fit_transform(df['processed_lemmatized'])

# Shape info
print(f'BoW matrix shape: {X_bow.shape}')
print(f'-> {X_bow.shape[0]} documents x {X_bow.shape[1]} features')

# Vocabulary
vocab = count_vectorizer.get_feature_names_out()

print(f'\nVocabulary size: {len(vocab)}')
print(f'First 20 features: {vocab[:20].tolist()}')
```

```python
tfidf_vectorizer = TfidfVectorizer(max_features=5000)

X_tfidf = tfidf_vectorizer.fit_transform(df['processed_lemmatized'])

print(f'\nTF-IDF matrix shape: {X_tfidf.shape}')
```

```python
tfidf_vocab = tfidf_vectorizer.get_feature_names_out()

doc_idx = 1
doc_vector = X_tfidf[doc_idx]

# Non-zero elements
nonzero_indices = doc_vector.nonzero()[1]

scores = [(tfidf_vocab[i], doc_vector[0, i]) for i in nonzero_indices]

# Sort by score
scores_sorted = sorted(scores, key=lambda x: x[1], reverse=True)

print(f'\nTop 10 TF-IDF terms for doc {doc_idx}:')
for term, score in scores_sorted[:10]:
    print(f'{term:<25} {score:.4f}')
```

```python
#PART3

import pandas as pd
import numpy as np
import string
import nltk

from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report

# Download required resources
nltk.download('punkt', quiet=True)
nltk.download('punkt_tab', quiet=True)
nltk.download('stopwords', quiet=True)
nltk.download('wordnet', quiet=True)

# Initialize tools
stop_words = set(stopwords.words('english'))
punct = set(string.punctuation)
lemmatizer = WordNetLemmatizer()

print('Setup complete.')
```

```python
# Using processed_lemmatized from Part 2
# Ensure labels numeric (required for sklearn models)

# Convert labels (positive/negative -> 1/0)
if df['label'].dtype == 'object':
    df['label'] = df['label'].map({'positive': 1, 'negative': 0})

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(
    df['processed_lemmatized'],
    df['label'],
    test_size=0.2,
    random_state=42
)

print(f'Train size: {len(X_train)}, Test size: {len(X_test)}')
```

```python
# Initialize TF-IDF vectorizer
tfidf = TfidfVectorizer(max_features=5000)

# Fit ONLY on training data
X_train_vec = tfidf.fit_transform(X_train)

# Transform test data using same vocabulary
X_test_vec = tfidf.transform(X_test)

print(f'\nTrain matrix shape: {X_train_vec.shape}')
print(f'Test matrix shape:  {X_test_vec.shape}')

print('Vectorizer fitted ONLY on training data (no data leakage).')
```

```python
# Initialize and train model (NB)
nb_model = MultinomialNB()
nb_model.fit(X_train_vec, y_train)

# Predictions
nb_preds = nb_model.predict(X_test_vec)
```

```python
print('=== Naive Bayes ===')

nb_accuracy = accuracy_score(y_test, nb_preds)
print(f'Accuracy: {nb_accuracy:.4f}')

print('\nClassification Report:')
print(classification_report(
    y_test,
    nb_preds,
    target_names=['negative', 'positive']
))
```

```python
#PART4

from nltk.util import ngrams
from nltk.probability import FreqDist

# Build corpus from training data (already preprocessed, cleaned)
train_corpus = ' '.join(X_train.tolist())

# Tokenize
train_tokens = train_corpus.split()

print(f'Training corpus: {len(train_tokens):,} tokens')
print(f'Sample tokens: {train_tokens[:20]}')
```

```python
# Unigram and Bigram frequency distributions
unigram_fd = FreqDist(train_tokens)
bigram_fd = FreqDist(ngrams(train_tokens, 2))

# Stats
vocab_size = len(unigram_fd)
total_tokens = sum(unigram_fd.values())

print(f'\nVocabulary size: {vocab_size:,}')
print(f'Total tokens: {total_tokens:,}')
print(f'Total bigram types: {len(bigram_fd):,}')

print('\nMost common unigrams:')
print(unigram_fd.most_common(10))

print('\nMost common bigrams:')
print(bigram_fd.most_common(10))
```

```python
def calculate_bigram_prob(word1, word2, unigram_fd, bigram_fd):
    """
    MLE: P(word2 | word1) = C(word1, word2) / C(word1)
    """
    count_w1 = unigram_fd[word1]

    if count_w1 == 0:
        return 0.0

    count_bigram = bigram_fd[(word1, word2)]

    return count_bigram / count_w1
```

```python
test_pairs = [('film', 'good'), ('bad', 'movie'),
              ('great', 'acting'), ('the', 'film')]

print(f"{'Word1':<15} {'Word2':<15} {'P(w2|w1)':<12} {'C(w1,w2)':<10} {'C(w1)'}")
print('-' * 60)

for w1, w2 in test_pairs:
    p = calculate_bigram_prob(w1, w2, unigram_fd, bigram_fd)
    print(f'{w1:<15} {w2:<15} {p:<12.6f} {bigram_fd[(w1,w2)]:<10} {unigram_fd[w1]}')
```

```python
def sentence_probability(sentence, unigram_fd, bigram_fd, total_tokens):
    """
    Bigram chain rule:
    P(w1,...,wn) = P(w1) * P(w2|w1) * ... * P(wn|wn-1)
    """
    tokens = sentence.lower().split()

    if not tokens:
        return 0.0

    # First word probability
    prob = unigram_fd[tokens[0]] / total_tokens

    # Multiply bigram probabilities
    for i in range(1, len(tokens)):
        p_bi = calculate_bigram_prob(tokens[i-1], tokens[i],
                                     unigram_fd, bigram_fd)
        if p_bi == 0:
            return 0.0
        prob *= p_bi

    return prob
```

```python
test_sentences = [
    "film good acting",
    "elephant cinema unseen"
]

print('\nSentence probabilities (UNSMOOTHED):')
for sent in test_sentences:
    p = sentence_probability(sent, unigram_fd, bigram_fd, total_tokens)
    print(f'P("{sent}") = {p:.2e}')
```

**Lemmatization** keeps semantic meaning but stemming cuts words aggressively, producing non-words sometimes. Lemmatization keeps context, better for ML models. Stemming reduces vocabulary size, sometimes helpful, but less precise.
In other words, Model is connecting tfidf to label sentiment/0,1. We need good sentiment connection. Good sentiment connection needs meaningfulness. Stem makes "better" to "bet". We lose meaning. Lemmatization would do better as it did.
