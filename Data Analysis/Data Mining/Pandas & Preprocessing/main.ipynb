#### Question 2. Explain how the DSM is adaptive.

DSM is designed and structured in such phases which allow it to be adaptive and flexible as there can be new information or data may arise. Each phase of
DSM is adaptive. First phase of problem understanding can have a certain understandings of objectives but as the project continues new developments or insights arise which allow for adjustments and redefine objectives. This continous refinement ensures that project evolves in response to emerging insights and be alignedw ith its goal.
For example, suppose business initially aims to predict a customer retension using existing customer data but team realizes that certain key features of customer interaction history were missing and not recorded. DSM's adaptive nature allows them to go back to data preparation phase and gather more data and re-engineer some features. Similarly same things can be done at each phase to produce better results.







#### Question 3. Describe how the DSM is iterative.

DSM is iterative because it involves repeated cycles of refinement througout its phases. Each phase is designed to inform and improve the previous phase which allows ongoing adjustments and enhancements. As new insights come to light, we have to take steps that are adaptive as we discussed in earlier answer. This makes DSM iterative.
For example, A team is buidling a recommendation system. After initial model is developed, they find its accuracy is not what was expected. This forces the team to return to their data preparation phase to add or create new features, then iterate throught the modeling and evaluation phase again. This cycle repeats until the model matches team's expectation which shows the iterative nature of DSM.







#### Question 4. Tuples/records with missing values for some attributes are common in real-world data. Describe various methods for handling this problem.

->Missing values are a common problem while processing the datas. Factors like equipment malfunction, human error, inconsistent record play a major role for the missing of the data. Now when we intend to process data for data mining we dont want missing data in our datasets and therefore we should solve this problem beforehand and then move towards further step of data mining. The various steps that can be employed to handle the missing data are:

1. Fill in the missing values manually: In this step the data miner can dig through the exisiting records, files and manually try to enter the data but while working in a large datasets this process becomes tidious and highly infeasible.

2. Introducing a global constant: A global constant can be introduced as a seperate category of data for all the missing values of the data in the datasets.

3. Using the atrribute mean: The missing data can be filled with the average of values present in the same attribute. Doing so we assume the missing data to fall into similar range of attributes and can be processed to get accurate result.

4. Using Inference: We can use the inference based on Bayesian network or decision tree to estimate the msising data. It solely depends on the concise of the data miner to put suitable value in missing entity.

5. Linear regression: Its a good method to predict the missing value of data. With linear regression we find a correlated value and see how the given attribtue is changing according to that value and then with the help of that we can predict the missing value.

These are the ways we can handle missing data in datasets.
