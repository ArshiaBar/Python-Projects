```python
# It's not valid. We could normalize the data, making everything out of 100grams for example.

def normalize_nutrition(food):
    factor = 100 / food["weight_in_grams"]   #providing factor
    food['saturated_fat']*=factor            #making other columns correspondent to 100grams of food
    food['cholesterol']*=factor
    return food

data=data.apply(normalize_nutrition, axis=1) #applying function
data=data.drop(columns=['weight_in_grams']) #all are 100grams, so the column is trivial
print(data)
```

**Task:** Derive a new variable, `saturated_fat_per_gram`, by dividing the amount of saturated fat by the weight in grams. Make sure you add the new variable to the `nutrition_subset` DataFrame.

**Task:** Sort the data set by `saturated_fat_per_gram` and produce a listing of the ten food items highest in saturated fat per gram.

```python
#using original data
data=pd.read_csv('gdrive/MyDrive/WS3/ws3data/nutrition_subset')

data['saturated_fat_per_gram'] = data['saturated_fat'] / data['weight_in_grams'] #new column
print(data.head)

sorted=data.sort_values(['saturated_fat_per_gram'],ascending=False)  #sorting with descending format
print(sorted[['food item','saturated_fat_per_gram']].head(10))
```

**Task:** Which food has the most saturated fat per gram?

```python
print(sorted['food item'].iloc[0]) #first food in sorted sequence
```

**Task:** Derive a new variable, `cholesterol_per_gram`. Make sure you add the new variable to the `nutrition_subset` DataFrame.

**Task:** Sort the data set by `cholesterol_per_gram` and then list the ten food items highest in cholesterol fat per gram.


```python
data['cholesterol_per_gram'] = data['cholesterol'] / data['weight_in_grams'] #new column
print(data.head)

sorted=data.sort_values(['cholesterol_per_gram'],ascending=False) #descending
print(sorted[['food item','cholesterol_per_gram']].head(10))
```

**Task:** Which food has the most cholesterol fat per gram?

```python
print(sorted['food item'].iloc[0]) #first food in sorted sequence
```

**Task:** Standardize the field `saturated_fat_per_gram`. Produce a listing of all the food items that are outliers at the high end of the scale. Use this condition to see the record at the high end `saturated_fat_per_gram_z>3`

```python
from scipy import stats
data['saturated_fat_per_gram_z'] = stats.zscore(data['saturated_fat_per_gram']) #standardizing
data_outliers = data.query('saturated_fat_per_gram_z>3')                        #upper outliers
print(data_outliers[['food item','saturated_fat_per_gram_z']].head)
```

**Task:** Based on the previous step, how many food items are outliers at the low end of the scale? Use this condition to see the record at the low end `saturated_fat_per_gram_z<-3`

```python
data_outliers_num = len(data.query('saturated_fat_per_gram_z<-3')) #quantity
print(data_outliers_num)
```

**Task:** Standardize the field `cholesterol_per_gram`. Produce a listing of all the food items that are outliers at the high end of the scale. Use this condition to see the record at the high end `cholesterol_per_gram >3`

```python
data['cholesterol_per_gram_z'] = stats.zscore(data['cholesterol_per_gram']) #standardize
data_outliers = data.query('cholesterol_per_gram_z>3')                      #upper outliers
print(data_outliers[['food item','cholesterol_per_gram_z']].head)
```

**Question 1**.	Why is it important to standardize variables like saturated_fat_per_gram and cholesterol_per_gram when comparing food items of different sizes?



Standardizing variables like saturated_fat_per_gram and cholesterol_per_gram is crucial when comparing food items of different sizes because it provides a consistent basis for evaluating nutrient content relative to the food's weight. Without standardization, larger food portions may appear to have higher nutrient values simply due to size, masking whether the food is genuinely higher in a particular nutrient per unit weight. By expressing nutrients on a per-gram basis, we can make fair comparisons between items of varying sizes, revealing true differences in nutrient density and enabling more accurate health assessments and dietary decisions.

**Question 2.**	Explain the process and reasoning behind identifying outliers in the dataset. Why is it useful to identify these outliers?


Below are the steps employed in identifying outlier in given dataset:
Process of Identifying Outliers

**Standardization**: The first step in identifying outliers is standardizing the saturated_fat_per_gram and cholesterol_per_gram values using Z-scores. This process transforms the values into a common scale, allowing for the identification of how far each value deviates from the mean in terms of standard deviations.

**Setting a Threshold**: By defining a threshold (e.g., Z-score > 3), we can classify certain food items as outliers. This threshold indicates that the food items in question have nutrient concentrations that are significantly higher than what would typically be expected.

**Filtering the Data:** Once the Z-scores are calculated, we filter the dataset to extract food items that exceed this threshold. This provides a focused view of the extreme cases that stand out due to their high levels of saturated fat or cholesterol per gram.

 This reason behind identifying these outliers in our dataset is for the identification of potential data entry errors or inconsistencies and to enhance the overall integrity of the dataset.So, often the entire result of data analysis result gets corrupted by the presence of the outliers. By identifying these outliers we can check if the given data was correct or caused by some human or technical error. Also, Identifying outliers in the nutritional dataset, specifically for saturated_fat_per_gram and cholesterol_per_gram, is essential for deriving accurate insights into food quality and health implications. Outliers can indicate food items with exceptionally high concentrations of these nutrients, which may signal unhealthy choices that require attention in dietary assessments. By using Z-scores to standardize and determine these outliers, we ensure that our analysis reflects meaningful deviations from typical values, enabling nutritionists and health professionals to focus on items that may adversely affect health outcomes.

Identifying outlier is useful as it helps to get accurate result in the data analysis process. It helps to get a consistent result and help use cross check the integrity of the data and if the data is correct then in above case it helps to highlight food items with significantly high levels of saturated fat or cholesterol per gram. This will suggest that such food would be harmful to the body and give the analyst idea about which are the food to avoid for better health.
