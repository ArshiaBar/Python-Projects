```python
import cv2

# Open the default camera
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Cannot open camera")
    exit()

# Get the width and height of the video capture
width = int(cap.get(cv2.CAP_PROP_FRAME_WIDTH))
height = int(cap.get(cv2.CAP_PROP_FRAME_HEIGHT))
print(f"Video resolution: {width}x{height}")

while True:
    # Capture frame-by-frame
    ret, frame = cap.read()

    # If frame is read correctly ret is True
    if not ret:
        print("Can't receive frame (stream end?). Exiting ...")
        break

    # Display the resulting frame
    cv2.imshow('Webcam Stream', frame)

    # Break the loop on 'q' key press
    if cv2.waitKey(1) == ord('q'):
        break

# Release the capture and close windows
cap.release()
cv2.destroyAllWindows()
```

Import OpenCV:
cv2 is the OpenCV library used for computer vision tasks.

Open the webcam:
cv2.VideoCapture(0) opens the default webcam (device index 0).

Check if camera opened successfully:
If fails, program exits with an error message.

Get video resolution:
It retrieves the width and height of the video stream using cv2.CAP_PROP_FRAME_WIDTH and cv2.CAP_PROP_FRAME_HEIGHT.

Start a loop to read frames:
cap.read() captures each frame.
If the frame is read successfully, it displays the frame in a window titled "Webcam Stream".

Exit on key press:
Pressing 'q' key breaks the loop and stops the video stream.

Release resources:
camera is released and all OpenCV windows are closed.

FPS Change to 10:


solution1: Set FPS using OpenCV's CAP_PROP_FPS
Requested FPS: 10, Actual FPS: 15.0

Not reliable across all systems.
Many webcams ignore FPS settings from OpenCV.
Works better with video files than live camera feeds.

solution2: Manually control frame rate using time.sleep()
The video feed slows down to ~10 FPS (choppier motion)

Artificial delay may cause lag or frame skipping.
Doesn't reduce actual capture rate. just slows display and processing.

```python
#MODIFICATION

import cv2
import os

# Initialize webcam
cap = cv2.VideoCapture(0)

# Create a counter for image filenames
image_counter = 1

# Create output folder if needed
output_folder = "snapshots"
os.makedirs(output_folder, exist_ok=True)

while True:
    ret, frame = cap.read()
    if not ret:
        print("Can't receive frame. Exiting...")
        break

    cv2.imshow('Webcam Stream', frame)

    key = cv2.waitKey(1)

    # Exit on 'q'
    if key == ord('q'):
        break

    # Take snapshot on 'x'
    if key == ord('x'):
        # Crop 15 pixels from all sides
        cropped = frame[15:-15, 15:-15]

        # Pad with 40 pixels (choose replicate or constant)
        # Option 1: Replicate padding
        padded = cv2.copyMakeBorder(cropped, 40, 40, 40, 40, cv2.BORDER_REPLICATE)

        # Option 2: Constant padding (e.g., blue)
        # padded = cv2.copyMakeBorder(cropped, 40, 40, 40, 40, cv2.BORDER_CONSTANT, value=(255, 0, 0))

        # Generate filename
        filename = f"image{image_counter:02d}.jpg"
        filepath = os.path.join(output_folder, filename)

        # Save image
        cv2.imwrite(filepath, padded)
        print(f"Saved snapshot: {filename}")

        image_counter += 1

cap.release()
cv2.destroyAllWindows()
```
