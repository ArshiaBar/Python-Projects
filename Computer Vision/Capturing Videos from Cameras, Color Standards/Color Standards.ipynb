```python
import cv2
import numpy as np

# Load the image
image = cv2.imread("imagee.jpg")
if image is None:
    print("Error: imagee.jpg not found.")
    exit()

# Display original image
cv2.imshow("Original Image", image)
cv2.waitKey(0)
cv2.destroyAllWindows()

# Convert to HSV
hsv = cv2.cvtColor(image, cv2.COLOR_BGR2HSV)
h, s, v = cv2.split(hsv)

# --- Step 1: Add 50 to V channel ---
v_plus_50 = cv2.add(v, 50)
hsv_v = cv2.merge([h, s, v_plus_50])
bgr_v = cv2.cvtColor(hsv_v, cv2.COLOR_HSV2BGR)
cv2.imwrite("image_v_plus_50.jpg", bgr_v)
cv2.imshow("image_v_plus_50", bgr_v)
cv2.waitKey(0)
cv2.destroyAllWindows()

# --- Step 2: Double the S channel ---
s_double = cv2.multiply(s, 2)
s_double = np.clip(s_double, 0, 255).astype(np.uint8)
hsv_s = cv2.merge([h, s_double, v])
bgr_s = cv2.cvtColor(hsv_s, cv2.COLOR_HSV2BGR)
cv2.imwrite("image_s_double.jpg", bgr_s)
cv2.imshow("image_s_double", bgr_s)
cv2.waitKey(0)
cv2.destroyAllWindows()

# --- Step 3: Set H to 0 ---
h_zero = np.zeros_like(h)
hsv_h0 = cv2.merge([h_zero, s, v])
bgr_h0 = cv2.cvtColor(hsv_h0, cv2.COLOR_HSV2BGR)
cv2.imwrite("image_h_zero.jpg", bgr_h0)
cv2.imshow("image_h_zero", bgr_h0)
cv2.waitKey(0)
cv2.destroyAllWindows()

# --- Step 4: Set H to 90 ---
h_ninety = np.full_like(h, 90)
hsv_h90 = cv2.merge([h_ninety, s, v])
bgr_h90 = cv2.cvtColor(hsv_h90, cv2.COLOR_HSV2BGR)
cv2.imwrite("image_h_90.jpg", bgr_h90)
cv2.imshow("image_h_90", bgr_h90)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
