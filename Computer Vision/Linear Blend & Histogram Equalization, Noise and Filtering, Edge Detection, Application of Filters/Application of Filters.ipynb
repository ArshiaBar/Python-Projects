```python
import cv2
import numpy as np
import os

# Create output folder
output_folder = 'output_images'
os.makedirs(output_folder, exist_ok=True)

# Load the images
img1 = cv2.imread('images/Cat.jpeg')  
img2 = cv2.imread('images/Dog.jpeg')  

# Check if images loaded successfully
if img1 is None or img2 is None:
    raise FileNotFoundError("One or both images not found. Please check the paths.")

# Resize both images to 600x800 and convert to grayscale
target_size = (800, 600)
gray1 = cv2.cvtColor(cv2.resize(img1, target_size), cv2.COLOR_BGR2GRAY)
gray2 = cv2.cvtColor(cv2.resize(img2, target_size), cv2.COLOR_BGR2GRAY)

# Apply Gaussian blur with kernel size 101x101
blur1 = cv2.GaussianBlur(gray1, (101, 101), 0)
blur2 = cv2.GaussianBlur(gray2, (101, 101), 0)

# Show and save blurred images
cv2.imshow('Blurred Cat Image', blur1)
cv2.waitKey(0)
cv2.imwrite(os.path.join(output_folder, 'blurred_cat.jpg'), blur1)

cv2.imshow('Blurred Dog Image', blur2)
cv2.waitKey(0)
cv2.imwrite(os.path.join(output_folder, 'blurred_dog.jpg'), blur2)

# Subtract blurred Cat image from original Cat image
subtracted = cv2.subtract(gray1, blur1)
cv2.imshow('Subtracted Image (Cat - Blurred Cat)', subtracted)
cv2.waitKey(0)
cv2.imwrite(os.path.join(output_folder, 'subtracted_cat.jpg'), subtracted)

# Add blurred Dog image to the result
added = cv2.add(subtracted, blur2)
cv2.imshow('Final Added Image (Subtracted + Blurred Dog)', added)
cv2.waitKey(0)
cv2.imwrite(os.path.join(output_folder, 'final_added_image.jpg'), added)

# Display the final image at multiple scales
scales = [1.0, 0.8, 0.6, 0.4, 0.2]

for scale in scales:
    resized = cv2.resize(blur1, (0, 0), fx=scale, fy=scale)
    title = f'Final Image at {int(scale*100)}%'
    cv2.imshow(title, resized)
    cv2.waitKey(0)
    filename = f'final_image_{int(scale*100)}percent.jpg'
    cv2.imwrite(os.path.join(output_folder, filename), resized)


for scale in scales:
    resized = cv2.resize(blur2, (0, 0), fx=scale, fy=scale)
    title = f'Final Image at {int(scale*100)}%'
    cv2.imshow(title, resized)
    cv2.waitKey(0)
    filename = f'final_image_{int(scale*100)}percent.jpg'
    cv2.imwrite(os.path.join(output_folder, filename), resized)


for scale in scales:
    resized = cv2.resize(subtracted, (0, 0), fx=scale, fy=scale)
    title = f'Final Image at {int(scale*100)}%'
    cv2.imshow(title, resized)
    cv2.waitKey(0)
    filename = f'final_image_{int(scale*100)}percent.jpg'
    cv2.imwrite(os.path.join(output_folder, filename), resized)

for scale in scales:
    resized = cv2.resize(added, (0, 0), fx=scale, fy=scale)
    title = f'Final Image at {int(scale*100)}%'
    cv2.imshow(title, resized)
    cv2.waitKey(0)
    filename = f'final_image_{int(scale*100)}percent.jpg'
    cv2.imwrite(os.path.join(output_folder, filename), resized)

cv2.destroyAllWindows()
```
