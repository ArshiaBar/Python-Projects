```python
import cv2
import numpy as np
import os

# Create output folder
output_folder = 'output_images'
os.makedirs(output_folder, exist_ok=True)

# Load the image
img = cv2.imread('images/Credit.png') 

# Check if image loaded successfully
if img is None:
    raise FileNotFoundError("Image not found. Please check the path and try again.")

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Define Sobel parameters
sobel_params = [
    (1, 0, 'dx=1_dy=0'),
    (0, 1, 'dx=0_dy=1'),
    (1, 1, 'dx=1_dy=1')
]

# Apply Sobel operator with both CV_8U and CV_64F
for dx, dy, label in sobel_params:
    # CV_8U
    sobel_8u = cv2.Sobel(gray, cv2.CV_8U, dx, dy, ksize=3)
    title_8u = f'Sobel_{label}_CV_8U'
    cv2.imshow(title_8u, sobel_8u)
    cv2.waitKey(0)
    cv2.imwrite(os.path.join(output_folder, f'{title_8u}.jpg'), sobel_8u)

    # CV_64F
    sobel_64f = cv2.Sobel(gray, cv2.CV_64F, dx, dy, ksize=3)
    sobel_64f_abs = cv2.convertScaleAbs(sobel_64f)
    title_64f = f'Sobel_{label}_CV_64F'
    cv2.imshow(title_64f, sobel_64f_abs)
    cv2.waitKey(0)
    cv2.imwrite(os.path.join(output_folder, f'{title_64f}.jpg'), sobel_64f_abs)

# Close all OpenCV windows
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import os

# Create output folder
output_folder = 'output_images'
os.makedirs(output_folder, exist_ok=True)

# Load the image
img = cv2.imread('images/Credit.png')

# Check if image loaded successfully
if img is None:
    raise FileNotFoundError("Image not found. Please check the path and try again.")

# Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# Apply Canny edge detection with two threshold pairs
canny_1 = cv2.Canny(gray, 50, 150)
cv2.imshow('Canny Edge (50,150)', canny_1)
cv2.waitKey(0)
cv2.imwrite(os.path.join(output_folder, 'canny_50_150.jpg'), canny_1)

canny_2 = cv2.Canny(gray, 100, 200)
cv2.imshow('Canny Edge (100,200)', canny_2)
cv2.waitKey(0)
cv2.imwrite(os.path.join(output_folder, 'canny_100_200.jpg'), canny_2)

# Close all OpenCV windows
cv2.destroyAllWindows()
```

| Feature              | Sobel                              | Canny                                  |
|----------------------|------------------------------------|----------------------------------------|
| Type                 | Gradient-based                     | Multi-stage edge detection             |
| Output               | Gradient magnitude                 | Binary edge map                        |
| Directional control  | Yes (dx/dy)                        | No (automatic edge tracing)            |
| Noise sensitivity    | High                               | Low (includes noise reduction)         |
| Edge connectivity    | Poor (isolated edges)              | Good (connected edges via hysteresis)  |
| Thresholding         | Not built-in                       | Built-in (low/high thresholds)         |
